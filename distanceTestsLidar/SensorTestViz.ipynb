```python
import os
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
folder_names = [
    "testDataBox",
    "testDataTable",
    "testDataWhiteWall"
]
```

```python
for folder in folder_names:
    
    subfolders = []
    for filename in os.listdir(folder):
        subfolders.append(int(filename.strip(".txt")))
    subfolders.sort()
    
    current_bar = []
    current_files = []
    for int_file in subfolders:
        
        filename = str(int_file) + ".txt"
        
        f = os.path.join(folder, filename)
        if os.path.isfile(f):
            current_data = []
            f = open(f, "r")
            for x in f:
                if len(eval(x)) == 0:
                    current_data.append(0)
                else:
                    for tup in eval(x):
                        current_data.append(tup[2])
                    
            current_bar.append(sum(current_data)/len(current_data))
            current_files.append(filename.strip(".txt"))

            
    fig, ax = plt.subplots()
    ax.bar(current_files, current_bar)

    ax.set_ylabel('Sensor value')
    ax.set_xlabel('Distance')
    ax.set_title(folder)
    ax.legend(title='Sensor value')

    plt.show()
```
